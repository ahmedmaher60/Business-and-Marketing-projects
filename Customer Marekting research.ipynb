#Cleaning and processing

```python
#convert categorical to numerical
from sklearn.preprocessing import OrdinalEncoder
# 'categorical_columns' is a list of the categorical column names
categorical_columns = ['Education', 'Marital_Status', 'Dt_Customer']
ordinal_encoder = OrdinalEncoder()
df[categorical_columns] = ordinal_encoder.fit_transform(df[categorical_columns])
```

```python
#filling missing value with mean
# Impute missing values with column means
df = df.fillna(df.mean())
```

```python
num_cols = ['ID', 'Year_Birth', 'Education', 'Marital_Status', 'Income', 'Kidhome',
                'Teenhome', 'Dt_Customer', 'Recency', 'MntWines', 'MntFruits',
                'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
                'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
                'NumStorePurchases', 'NumWebVisitsMonth', 'AcceptedCmp3',
                'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2',
                'Complain', 'Z_CostContact', 'Z_Revenue', 'Response']
```

# Features Engineering

```python
X = df.drop(columns=['Response'])  # Replace 'TargetColumn' with the actual target column name
y = df['Response']
```

```python

from sklearn.feature_selection import mutual_info_classif
mutual_info = mutual_info_classif(X, y)
```

```python
mi_scores = pd.Series(mutual_info, index=X.columns)
mi_scores = mi_scores.sort_values(ascending=False)  # Sort by descending mutual information scores
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
mi_scores.plot(kind='bar')
plt.title("Mutual Information Scores")
plt.ylabel("Mutual Information")
plt.show()
```

```python
# Display the top N features with their MI scores
top_N = 24
selected_features = mi_scores.head(top_N).index.tolist()
print("Selected Features:")
print(selected_features)
```

```python
N_X = df[selected_features]
```

```python
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyClassifier

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(N_X, y, test_size=0.2, random_state=42)
```

```python

# Initialize LazyClassifier
clf = LazyClassifier(predictions=True)

# Fit and predict using LazyClassifier
models = clf.fit(X_train, X_test, y_train, y_test)

# Print the models summary
print(models)
```

```python
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMClassifier
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    # Add more hyperparameters to tune
}
```

```python
lgbm_model = LGBMClassifier()
grid_search = GridSearchCV(lgbm_model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
best_lgbm_model = grid_search.best_estimator_
y_pred = best_lgbm_model.predict(X_test)
print("Best Hyperparameters:", grid_search.best_params_)
print(classification_report(y_test, y_pred))
```
